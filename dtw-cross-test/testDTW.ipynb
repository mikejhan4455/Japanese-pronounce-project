```python
def draw_pitch(pitch):
    # Extract selected pitch contour, and
    # replace unvoiced samples by NaN to not plot
    pitch_values = pitch.selected_array['frequency']
    #print(pitch_values)
    pitch_values[pitch_values==0] = np.nan
    plt.plot(pitch.xs(), pitch_values,markersize=5, color='w')
    plt.plot(pitch.xs(), pitch_values,markersize=2)
    plt.grid(False)
    plt.ylim(0, pitch.ceiling)
    plt.ylabel("fundamental frequency [Hz]")
```

```python
import csv
import numpy as np
def v_pitch(audio):
    snd = parselmouth.Sound(audio)
    pitch = snd.to_pitch()
    #pitchs = snds.to_pitch()
    pitch_values = pitch.selected_array['frequency']
    time_pitch = pitch.xs()
    list_2 = pitch_values.tolist()
    list_1 = time_pitch.tolist()
    dataset = list(zip(list_1,list_2))
    return(dataset)
```

```python
import csv
import numpy as np
def v_intensity(audio):
    snd = parselmouth.Sound(audio)
    intensity = snd.to_intensity()
    list_1 = intensity.xs()
    list_2 = intensity.values.T
    dataset = list(zip(list_1,list_2))
    return(dataset)
```
